# EDA

## Percentage of Total Customer Churn and No Churn

```python
churn = data_train[data_train['churn'] == 'yes']
not_churn = data_train[data_train['churn'] == 'no']

# Mengubah nilai 'Yes' menjadi 1 dan 'No' menjadi 0 dalam kolom 'churn'
data_train['churn'] = data_train['churn'].replace({'Yes': 1, 'No': 0})

print("Churn: %i (%.1f%%)" % (len(churn), float(len(churn)) / len(data_train) * 100.0))
print("Not Churn: %i (%.1f%%)" % (len(not_churn), float(len(not_churn)) / len(data_train) * 100.0))
print("Total: %i" % len(data_train))
```

```python
import matplotlib.pyplot as plt

# Mengatur tema ggplot
plt.style.use('ggplot')

# Menghitung jumlah data untuk setiap kategori ('Churn' dan 'Not Churn')
churn_count = len(churn)
not_churn_count = len(not_churn)

# Membuat data untuk pie chart
sizes = [churn_count, not_churn_count]
labels = ['Customer Churn', 'Customer Not Churn']
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # Untuk memberikan efek "explode" pada bagian 'Churn'

# Membuat pie chart
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)

# Menambahkan judul
plt.title('Percentage Customer Churn vs Customer Not Churn', color='black')

# Menampilkan pie chart
plt.show()
```

##Detect Outlier

```python
#membuat variabel baru
numerical_data=['account_length', 'number_vmail_messages', 'total_day_minutes',
       'total_day_calls', 'total_day_charge', 'total_eve_minutes',
       'total_eve_calls', 'total_eve_charge', 'total_night_minutes',
       'total_night_calls', 'total_night_charge', 'total_intl_minutes',
       'total_intl_calls', 'total_intl_charge',
       'number_customer_service_calls']
```

```python
list(numerical_data)
```

```python
#handling outliers data train
print(f'jumlah baris sebelum memfilter outlier: {len(data_train)}')
filtered_entries = np.array([True] * len(data_train))
for col in numerical_data:
       Q1 = data_train[col].quantile(0.25)
       Q3 = data_train[col].quantile(0.75)
       IQR = Q3 - Q1
       low_limit = Q1 - (IQR * 1.5)
       high_limit = Q3 + (IQR * 1.5)
       filtered_entries = ((data_train[col] >= low_limit) & (data_train[col] <= high_limit)) &filtered_entries

data_train = data_train[filtered_entries]

print(f'jumlah baris setelah memfilter outlier: {len(data_train)}')
```

##Train-test Split

```python
from sklearn.model_selection import train_test_split
```

```python
#separating target variables from training features

x = data_train.drop(['churn'], axis = 1) #dropping the churn column as it is target variabels
y = data_train['churn']
```

```python
#split data into train and test data with 70:30 method
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape
```

##Feature Selection

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
#select the relevant feature
select = SelectKBest(score_func=chi2, k=10)
select.fit(x_train, y_train)

vector_names = list(x_train.columns[select.get_support(indices=True)])
print(vector_names)
```

##MODELLING DATA

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn import metrics
```

**Compare two Algorithms to choose which algorithm has the best accuracy to predict customer churn**

##KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#data classification KNN
knn = KNeighborsClassifier()
knn = knn.fit(x_train, y_train)
knn_predict = knn.predict(x_test)
knn_predicted = pd.Series(list(knn_predict))
knn_predicted.value_counts()
```

```python
print(accuracy_score(y_test, knn_predict))
```

```python
class_names=[-1,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="Blues" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix KNN', y=1.1)
plt.ylabel('Predicted label')
plt.xlabel('Actual label')
plt.show()
```

#EVALUATING MODEL

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

kfold = KFold(n_splits=10, shuffle=True, random_state=0)
```

##Evaluating Model KNN

```python
#Select best K
k_values = [i + 1 for i in range(20)]
accuracies = []

for k in k_values:
  model = KNeighborsClassifier(n_neighbors = k)
  results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
  accuracies.append((results.mean()*100.0, results.std()*100.0))
  print('k = ', k)
  print('Accuracy = %.3f%% (%.3f%%)' % (results.mean()*100.0, results.std()*100.0))

  best_k_index = accuracies.index(max(accuracies))
  best_k = k_values[best_k_index]
  best_accuracy = max(accuracies)

  print('Best K = ', best_k)
  print('Best Accuracy = ', best_accuracy)
```

##Evaluating Model Decision Tree

```python
dtc_model = DecisionTreeClassifier()
score = cross_val_score(dtc_model, x_train, y_train, cv=kfold, n_jobs=1, scoring='accuracy')

print('Accuracy = %.3f%% (%.3f%%)' % (score.max()*100.0, score.std()*100.0))
```
